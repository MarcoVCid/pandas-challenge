```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
completedf = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
completedf.dtypes
schools = len(completedf["School ID"].unique())
students = len(completedf['Student ID'].unique())
budget = school_data['budget'].sum()
math = completedf['math_score'].mean()
reading = completedf['reading_score'].mean()
passed_math = completedf[completedf['math_score'] >= 70]
per_math = len(passed_math['Student ID'])/students*100
passed_reading = completedf[completedf['reading_score'] >= 70]
per_reading = len(passed_reading['Student ID'])/students*100
passed_both = passed_math[passed_math['reading_score'] >= 70]
per_both = len(passed_both['Student ID'])/students*100

summary_df = pd.DataFrame({
    'Number of Schools': [schools],
    'Number of Students': [students],
    'Total Budget': [budget],
    'Average Math Score': [math],
    'Average Reading Score': [reading],
    '% Passing Math': [per_math],
    '% Passing Reading': [per_reading],
    '% Overall Passing': [per_both]
})

summary_df['Number of Students'] = summary_df['Number of Students'].map("{:,.0f}".format)
summary_df['Total Budget'] = summary_df['Total Budget'].map("${:,.2f}".format)
summary_df['Average Math Score'] = summary_df['Average Math Score'].map("{:,.2f}".format)
summary_df['Average Reading Score'] = summary_df['Average Reading Score'].map("{:,.2f}".format)
summary_df['% Passing Math'] = summary_df['% Passing Math'].map("{:,.2f}%".format)
summary_df['% Passing Reading'] = summary_df['% Passing Reading'].map("{:,.2f}%".format)
summary_df['% Overall Passing'] = summary_df['% Overall Passing'].map("{:,.2f}%".format)
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary = completedf[['school_name','type','Student ID','budget','math_score','reading_score']]
school_summary = school_summary.rename(columns={'school_name':'School Name','type':'School Type','Student ID':'Total Students',
                                               'budget':'Total Budget','math_score':'Average Math Score',
                                                'reading_score':'Average Reading Score'})
dist_df = school_summary[['School Name','School Type']]
math_df = school_summary[school_summary['Average Math Score'] >= 70]
math = math_df.groupby('School Name',as_index=False).count()
reading = school_summary[school_summary['Average Reading Score'] >= 70].groupby('School Name',as_index=False).count()

school_summary = school_summary.groupby('School Name',as_index=False)\
                            .agg({'Total Students': 'count', 'Total Budget':'mean', 'Average Math Score':'mean',
                                 'Average Reading Score':'mean'})

school_summary['% Passing Math'] = math['Average Math Score']/school_summary['Total Students']*100
school_summary['% Passing Reading'] = reading['Average Reading Score']/school_summary['Total Students']*100
school_summary['% Overall Passing'] = school_summary['% Passing Math']*school_summary['% Passing Reading']/100
school_summary['Per Student Budget'] = school_summary['Total Budget']/school_summary['Total Students']

dist_df = dist_df.drop_duplicates(subset=['School Name'])

school_summary = school_summary.merge(dist_df,on='School Name')

school_summary = school_summary[['School Name','School Type','Total Students','Total Budget','Per Student Budget',
                                 'Average Math Score','Average Reading Score',
                                '% Passing Math','% Passing Reading','% Overall Passing']]

school_summary2 = school_summary.set_index('School Name')

school_summary2['Total Students'] = school_summary2['Total Students'].map("{:,.0f}".format)
school_summary2['Total Budget'] = school_summary2['Total Budget'].map("${:,.2f}".format)
school_summary2['Average Math Score'] = school_summary2['Average Math Score'].map("{:,.2f}".format)
school_summary2['Average Reading Score'] = school_summary2['Average Reading Score'].map("{:,.2f}".format)
school_summary2['% Passing Math'] = school_summary2['% Passing Math'].map("{:,.2f}%".format)
school_summary2['% Passing Reading'] = school_summary2['% Passing Reading'].map("{:,.2f}%".format)
school_summary2['% Overall Passing'] = school_summary2['% Overall Passing'].map("{:,.2f}%".format)
school_summary2['Per Student Budget'] = school_summary2['Per Student Budget'].map("${:,.2f}".format)

school_summary2
```

```python
top_school = school_summary2.sort_values('% Overall Passing', ascending=False)
top_school = top_school.head(5)
top_school
```

```python
grade_m_df = completedf[['grade','school_name','math_score']]

bymath_grades = pd.DataFrame({'school name':[],
                 '9th':[],
                 '10th':[],
                 '11th':[],
                 '12th':[]})
index = 0

for name in grade_m_df['school_name'].drop_duplicates():
    maths = grade_m_df[grade_m_df['school_name'] == name].groupby('grade').mean()
    bymath_grades.loc[index] = [name,maths['math_score'].loc['9th'],maths['math_score'].loc['10th'],
                   maths['math_score'].loc['11th'],maths['math_score'].loc['12th']]
    index += 1
#nineth_m
bymath_grades = bymath_grades.sort_values('school name').set_index('school name')

bymath_grades['9th'] = bymath_grades['9th'].map('{:.2f}'.format)
bymath_grades['10th'] = bymath_grades['10th'].map('{:.2f}'.format)
bymath_grades['11th'] = bymath_grades['11th'].map('{:.2f}'.format)
bymath_grades['12th'] = bymath_grades['12th'].map('{:.2f}'.format)

bymath_grades
```

## Reading Score by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bymath = school_summary[['Per Student Budget','Average Math Score','Average Reading Score','% Passing Math',
                         '% Passing Reading','% Overall Passing']]

bins = [0,584.9,629.9,644.9,680]
group_names = ['<$585','$585-630','$630-645','$645-680']

bymath['Spending Ranges (Per Student)'] = pd.cut(bymath['Per Student Budget'], bins, labels=group_names)
bymath = bymath.groupby('Spending Ranges (Per Student)').mean()

bymath = bymath[['Average Math Score','Average Reading Score','% Passing Math',
                          '% Passing Reading','% Overall Passing']]

bymath['Average Math Score'] = bymath['Average Math Score'].map('{:.2f}'.format)
bymath['Average Reading Score'] = bymath['Average Reading Score'].map('{:.2f}'.format)
bymath['% Passing Math'] = bymath['% Passing Math'].map('{:.2f}%'.format)
bymath['% Passing Reading'] = bymath['% Passing Reading'].map('{:.2f}%'.format)
bymath['% Overall Passing'] = bymath['% Overall Passing'].map('{:.2f}%'.format)

bymath
```

## Scores by School Size
